```python
total_net_BrandRevenue =  total_retail_price - total_discounted_price
```

```python
brand_subset = pd.DataFrame(data= {'total_number_of_orders':total_number_of_orders,'total_retail_price':total_retail_price,'total_discounted_price':total_discounted_price,'total_net_BrandRevenue':total_net_BrandRevenue})
```

```python
brand_subset
```

```python
sns.pairplot(brand_subset)
```

## Q 9.Compare performance regionwise

## a)	Draw a lineplot for the monthly net Revenue of E-Commerce Company for each region separately.  (4 marks) 

## b)	Identify the best and the worst performing months for each region. (2 marks)

#### Note: Only those days with actual orders(Order_Date) placed are present in the dataset. Assuming there were no orders on other days

### Ans 9 a)

```python
import datetime
new_dates = []
for i in range(len(ECom_Data['Order_Date'])):
  dates = datetime.datetime.strptime(ECom_Data['Order_Date'][i],"%d-%m-%Y")
  new_dates.append(dates.month)
```

```python
new_dates = np.array(new_dates)
```

```python
new_dates.shape
```

```python
profitList = ECom_Data['retail_price'].tolist()
```

```python
profitList = ECom_Data['retail_price'].tolist()
plt.figure(figsize=(10,10))
sns.lineplot(new_dates,profitList,hue=ECom_Data['Region'])
```

### Ans 9 b)

```python
plt.figure(figsize=(100,100))
profitList = ECom_Data['retail_price'].tolist()
monthList  = new_dates
plot = sns.countplot(y= profitList,hue= new_dates)
```
